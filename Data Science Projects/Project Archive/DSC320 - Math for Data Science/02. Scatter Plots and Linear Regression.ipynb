## Assignment 1
Make a scatterplot of the highway miles per gallon (y-axis) versus the weight (x-axis)

```python
"""
Code Description: 

    - Creating a scatterplot to see the relationship between: 'hwy_mpg' vs 'weight'
    - Code Breakdown:
        - 1st Part: 
            - Loading 'car_data.csv' file
            - Extracting columns 'hwy_mpg' vs 'weight'
        - 2nd Part: 
            - Building a scatter plot via 'matplotlib'
"""

# 1st Part:
s_plot = pd.read_csv('car_data.csv')
weight = s_plot['weight']
hwy_mpg = s_plot['hwy_mpg']

# 2nd Part:
plt.figure(figsize=(8, 6))
plt.scatter(weight, hwy_mpg, alpha=0.7)
plt.title('Highway MPG vs. Weight')
plt.xlabel('Weight')
plt.ylabel('Highway MPG')
plt.grid(True)

# Printing plot
plt.show()
```
