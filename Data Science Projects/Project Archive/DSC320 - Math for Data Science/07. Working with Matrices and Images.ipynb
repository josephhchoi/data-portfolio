## Working with Matrices

```python
# Setup
import numpy as np
```

#### 1. Enter each of the matrices A, B, and C as Numpy matrix

```python
"""
Code Description:
    - Using NumPy to create matrices for A, B, C via "np.matrix()"

Code Breakdown:
    - 1st Part: Matrix A     
    - 2nd Part: Matrix B
    - 3rd Part: Matrix C
"""

# 1st Part:
a = np.matrix([[1, -2, 3, 7],
               [2, 1, 1, 4],
               [-3, 2, -2, 10]])

# 2nd Part:
b = np.matrix([[0, 1, -3, -2],
               [10, -1, 2, -3],
               [5, 1, -1, 4]])

# 3rd Part:
c = np.matrix([[4, 0, -2, 3],
               [3, 6, 9, 7],
               [2, 2, 5, 1],
               [9, 4, 6, -2]])
```

#### 2. What are the dimensions of each of the matrices A, B, and C? Display these using Python code

```python
"""
Code Description:
    - Displaying the dimensions of each matrices via "shape"

Code Breakdown:
    - 1st Part: Dimensions of Matrix A     
    - 2nd Part: Dimensions of Matrix B
    - 3rd Part: Dimensions of Matrix C
"""

# 1st Part:
a_dimensions = a.shape
print("A:", a_dimensions) # 3 rows and 4 columns

# 2nd Part:
b_dimensions = b.shape
print("B:", b_dimensions) # 3 rows and 4 columns

# 3rd Part:
c_dimensions = c.shape
print("C:", c_dimensions) # 4 rows and 4 columns
```

#### 3. Determine if each of the following matrix operations is defined. If it is defined, calculate the result using Python. If it is not defined, explain why it is not defined
**(a)** A + B <br>
**(b)** AB <br>
**(c)** AC <br>
**(d)** C^T <br>

```python
# "A + B" can be defined

# Calculating the resulting using Python:
result_3a = a + b
result_3a
```

```python
# "AB" cannot be defined; For multiplying matrices, the number of columns in A must be the same as the number of rows in B

# Calculating the resulting using Python (not defined):
result_3b = a * b
result_3b
```

```python
# "AC" can be defined

# Calculating the resulting using Python:
result_3c = a * c
result_3c
```

```python
# "C^T" can be defined (swapping matrix's rows and columns)

# Calculating the resulting using Python:
result_4c = np.transpose(c)
result_4c
```

#### 4. Illustrate the following property of matrix transpositions using these matrices: (A + B)^T = A^T + B^T
**(a)** Calculating "(A + B)^T" <br>
**(b)** Calculating "A^T + B^T" <br>
**(c)** Checking if the property of matrix transpositions is true via "np.array_equal" 

```python
# Calculating (A + B)^T

result_4a = np.transpose(a + b)
result_4a
```

```python
# Calculating A^T + B^T

result_4b = np.transpose(a) + np.transpose(b)
result_4b
```

```python
# Notes: Property states that the transpose of the sum of two matrices (A + B) is equal to the sum of their transposes (A^T + B^T). Is this true or false?

# Checking if the property of matrix transpositions is true via "np.array_equal"

result_4c = np.array_equal(result_4a, result_4b)
result_4c
```

#### 5. Illustrate the following property of matrix transposition using these matrices (AC)^T = C^T A^T
**(a)** Calculating "(AC)^T" <br>
**(b)** Calculating "C^T A^T" <br>
**(c)** Checking if the property of matrix transpositions is true via "np.array_equal" 

```python
# Calculating (AC)^T

result_5a = np.transpose(a*c)
result_5a
```

```python
# Calculating C^T A^T

result_5b = np.transpose(c) * np.transpose(a)
result_5b
```

```python
# Notes: Property states that the transpose of the product of two matrices (AC) is equal to the product of their transposes (C^T * A^T). Is this true or false?

# Checking if the property of matrix transpositions is true via "np.array_equal"

result_5c = np.array_equal(result_5a, result_5b)
result_5c
```

#### 6. Find C^-1 using Python code

```python
# Calculating inverse of C via "np.linalg.inv"

result_6 = np.linalg.inv(c)
result_6
```

## Working with Images

```python
# Setup
import cv2
import matplotlib.pyplot as plt
```

#### 1. Use the following code to display the image

```python
# Importing the image
image_bgr = cv2.imread('week7_image.jpg', cv2.IMREAD_COLOR)

# Converting the image to the RGB format
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# Displaying the image
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

#### 2.  Find the resolution of this image. Hint: OpenCV imports the image as a Numpy array, where each pixel corresponds to an entry in the array

```python
# Retrieving the resolution of the image via "shape"
resolution = image_rgb.shape[:2]

# Printing the results (height, width)
resolution
```

**Response**: The third number in the array is the number of color channels. In this case, our color channels are Red, Green, and Blue. The third number is part of the array because the image is using the RGB color model.  

```python
# Getting the color_channel of the image
color_channel = image_rgb.shape[2]

# Printing the results
color_channel
```
